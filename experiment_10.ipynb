```python
file_path = 'iris.csv'  # Replace with your dataset file path
df = pd.read_csv(file_path)
```

```python
print("Dataset Structure:")
print(df.info())
print("\nFirst 5 Rows of the Dataset:")
print(df.head())
print("\nColumn Names:")
print(df.columns.tolist())
```

```python
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numerical_cols].hist(figsize=(12, 8), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar', color='orange', figsize=(8, 4))
    plt.title(f"Bar Plot for {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in categorical_cols:
    df[col].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6), startangle=90)
    plt.title(f"Pie Chart for {col}")
    plt.ylabel("")  # Remove y-label for cleaner look
    plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    # Replace 'dataset' with 'df'
    sns.boxplot(data=df, y=col, color='lightgreen')
    plt.title(f"Box Plot for {col}")
    plt.ylabel(col)
    plt.show()
```

```python
cat_col = categorical_cols[0]  # Replace with your desired categorical column
for col in numerical_cols:
    plt.figure(figsize=(8, 6))
    # Replace 'dataset' with 'df' if that is the DataFrame containing the data.
    sns.violinplot(x=cat_col, y=col, data=df, palette='muted')
    plt.title(f"Violin Plot: {col} vs {cat_col}")
    plt.xlabel(cat_col)
    plt.ylabel(col)
    plt.show()
```

```python
sns.pairplot(df[numerical_cols])
plt.suptitle("Pairwise Relationships", y=1.02, fontsize=16)
plt.show()
```

```python
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```
